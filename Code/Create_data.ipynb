```python
#improt required package
import pandas as pd
import numpy as np
```

```python
#convert txt data to csv dataframe
path = "/Users/shenglienlee/Desktop/Work/2023_Fall/Deep_Learning/Project_3/Data/tick_data/0827/TSLA_30min.txt"
TSLA_30 = pd.read_csv(path, delimiter=',')
#add column name
TSLA_30.columns = ['Timestamp','Open','High','Low','Close','Volume']
#save as .csv file
TSLA_30.to_csv("TSLA_30min.csv",index=False)
```

```python
#2012-2017
TSLA_30['Timestamp'] =pd.to_datetime(TSLA_30['Timestamp'])
mask = (TSLA_30['Timestamp'] >= '2012-11-16') & (TSLA_30['Timestamp'] <= '2017-12-31')
TSLA_30_start_with_12 = TSLA_30.loc[mask].reset_index(drop=True)
TSLA_30_start_with_12.to_csv("TSLA_30min_2012.csv",index=False)
TSLA_30_start_with_12
```
